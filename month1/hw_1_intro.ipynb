```python
path = '.mnist'
```

```python
train_data = datasets.MNIST(path, train=True, download=True)
test_data  = datasets.MNIST(path, train=False, download=True)

train_mean = train_data.train_data.float().mean()/255
train_std  = train_data.train_data.float().std()/255

print(train_mean)
print(train_std)
```

```python
normalize_transform = transforms.Compose([
                transforms.ToTensor(),
                transforms.Normalize((train_mean,), (train_std,)),
           ])
```

```python
train_data = datasets.MNIST(path, train=True, download=True, transform=normalize_transform)
test_data = datasets.MNIST(path, train=False, download=True, transform=normalize_transform)
```
